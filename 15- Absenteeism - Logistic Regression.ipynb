## 1- Create the target
we will use a simple approach to deal with targets, we assume two classes: moderately absent and excessively absent and to define which one is which we take the median, if the abesnteeism time is less than median it is considered to be a moderatley absent case and is it is higher so we have the excessively absent case. 

#### Why using median?
1- because it is not affected by outliers

2- we implicitly balance the data sine half of our observations are below and half are above medium (as balancing prevent the model from learning to output only 0s or only 1s)

```python
# let's find the median
data_preprocessed['Absenteeism Time in Hours'].median()
```

```python
# Absenteeism Time in Hours <= 3 : moderately absent : return 0
# Absenteeism Time in Hours >= 4 : excessively absent: return 1
# inseted of For loop and IF we can use WHERE method
# np.where(condition, value if true, value if false)
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] <= data_preprocessed['Absenteeism Time in Hours'].median(), 0, 1)
targets
```

### How to interepret the odds_ratio?
As you see from the above table if coefficient is around zero it odds_ratio is around 1.
##### odds_ratio = for one unit change in the standardized feature, the odds increase by a multiple equal to the odds ratio. (odds_ratio = 1 => no change)
in the below formula if bk = 0 then e^bk = e^0 = 1 , so change in X (X2-X1) has no effect in Odds : (Odds2 - Odds1 = 0 )

$ log{\frac{P(X)}{1-P(X)}} = b_kX + b_0 =>$ 
$ log{\frac{odds_2}{odds_1}} = {{b_k}*{(X_2-X_1)}} =>$
$ \frac{odds_2}{odds_1} = e^{{b_k}*{(X_2-X_1)}} =>$
$ odds_2 = e^{{b_k}*{(X_2-X_1)}}*odds_1 $

if odds = 5:1 and odds_ratio (e^bk) = 2 then if x increases by 1 unit it odds inceases by 2 units : New odds = 10:1

for example the odds_ratio for transportation expense is 2, so if the tranportation expense increases by 1 unit (for standardized inputs 1 unit is kind of vague) , its odds for being excessively absent increases by 2 units. 
Features 6, 8, 10 have the coefficents of almost 0 and the odds ratio of 1, so they are not so important on determining the absence status of employee so we consider dropping them later on

```python
# import the relevant modules
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

# create the class which works similar to StandardScaler with only one difference
# StandardScaler standardizes all the inputs but our CustomScaler only standardzie the 
# the only columns that we choose. 
   
    # create the Custom Scaler class

class CustomScaler(BaseEstimator,TransformerMixin): 
    
    # init or what information we need to declare a CustomScaler object
    # and what is calculated/declared as we do
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        
        # scaler is nothing but a Standard Scaler object
        self.scaler = StandardScaler(copy,with_mean,with_std)
        # with some columns 'twist'
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    
    # the fit method, which, again based on StandardScale
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    # the transform method which does the actual scaling

    def transform(self, X, y=None, copy=None):
        
        # record the initial order of the columns
        init_col_order = X.columns
        
        # scale all features that you chose when creating the instance of the class
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        
        # declare a variable containing all information that was not scaled
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        
        # return a data frame which contains all scaled features and all 'not scaled' features
        # use the original order (that you recorded in the beginning)
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```
